```python
def plot_barplot(x, title, period=None, ax=None):
    if period is None:
        period = (x.max() - x.min()) / 20

    bins = list(np.arange(x.min(), x.max(), period))
    out = pd.cut(x, bins=bins)
    
    ax.set_title(title)
    out.value_counts(sort=False).plot.bar(color="b", ax=ax)
    
    return ax
#     image_name = os.path.join(IMAGES_DIR, '{}.png'.format(title))
#     plt.savefig(image_name)
```

```python
from tqdm import tqdm_notebook
import matplotlib.pyplot as plt
from IPython.display import clear_output
%matplotlib inline

axes_size = (5, 6)
fig, axes = plt.subplots(*axes_size, figsize=(81, 42))
ref_batch_name = 'GSE33828'

ref_data = train_data[train_data['GEO'] == ref_batch_name]
def shift_with_log(X):
    return np.log(X + 3)

for i, gene_name in tqdm_notebook(enumerate(best_genes)):
    i1 = int(i  % (axes_size[0] * axes_size[1]) / axes_size[1])
    i2 = int(i  % (axes_size[0] * axes_size[1]) % axes_size[1])
    x = ref_data[gene_name]
    x = shift_with_log(x)
    plot_barplot(x, 'barplot_{}'.format(gene_name), ax=axes[i1, i2])
    
    if (i + 1) % (axes_size[0] * axes_size[1]) == 0:
        plt.savefig('tmp_image.png')
        plt.show()
        input("Press Enter to continue...")
        print('here')
        plt.clf()
        fig, axes = plt.subplots(*axes_size, figsize=(81, 42))

```
