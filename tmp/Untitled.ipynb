```python
import json
```

```python
with open('cv_results_dae.json', 'r') as file:
    data1 = json.load(file)
    
with open('cv_results.json', 'r') as file:
    data2 = json.load(file)
```

```python
data_count1 = len(data1)
data_count2 = len(data2)

i_s = []
j_s = []
ij = []

for i in range(data_count1):
    _params = data1[i]['params']
#     print(_params)
    
    for j in range(data_count2):
        _params2 = data2[j]['params']
        
        if (_params == _params2):
#             if i != j:
#                 print()
            pair_name = '{}_{}'.format(i, j)
            reversed_pair_name = '{}_{}'.format(j, i)

            i_s.append(i)
            j_s.append(j)

            if pair_name not in ij and reversed_pair_name not in ij:
                ij.append(pair_name)
                
print(sorted(i_s))
print(sorted(j_s))
print(sorted(ij, key=lambda x: int(x.split('_')[0])))
```

```python
with open('cv_results_dae.json', 'r') as file:
    data2 = json.load(file)
    
with open('cv_results_dae_test.json', 'r') as file:
    data1 = json.load(file)
```

```python
data_count1 = len(data1)
data_count2 = len(data2)

i_s = []
j_s = []
ij = []

answer = []

for i in range(data_count1):
    _params = data1[i]['params']
    
    for j in range(data_count2):
        _params2 = data2[j]['params']
        
        if (_params == _params2):
            
#             print(data1[i]['model_path'])
            if 'trained_models' in data2[j]['model_path']:
                pair_name = '{}_{}'.format(i, j)
                reversed_pair_name = '{}_{}'.format(j, i)

                i_s.append(i)
                j_s.append(j)

                if pair_name not in ij and reversed_pair_name not in ij:
                    ij.append(pair_name)
                    
                write_message = dict(
                    params=data1[i]['params'],
                    scores=data2[j]['scores'],
                    model_path=data2[j]['model_path']
                )
                answer.append(write_message)

                
experiment_meta_params_file = 'cv_results_dae_kfold.json'
with open(experiment_meta_params_file, 'w') as file:
    json.dump(answer, file)
                
print(sorted(i_s))
print(sorted(j_s))
print(sorted(ij, key=lambda x: int(x.split('_')[0])))
```
